```python
import matplotlib.pyplot as plt

# Criar um gráfico de linha com rótulos de texto para cada ponto
plt.plot(df['Mês'], df['Raivas que passei'], marker='o', linestyle='-')
plt.xlabel('Mês')
plt.ylabel('Raivas que passei')
plt.title('Raivas que passei ao longo dos meses')
plt.xticks(rotation=45)
plt.grid(True)

# Adicionar rótulos de texto aos pontos
for i, txt in enumerate(df['Raivas que passei']):
    plt.text(df['Mês'][i], df['Raivas que passei'][i], str(txt), fontsize=9, ha='right', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Criar um gráfico hexbin para mostrar a densidade dos pontos
plt.hexbin(df['Mês'], df['Raivas que passei'], gridsize=20, cmap='Blues')
plt.xlabel('Mês')
plt.ylabel('Raivas que passei')
plt.title('Densidade de Raivas que passei ao longo dos meses')
plt.colorbar(label='Contagem de Raivas')
plt.grid(True)
plt.show()
```
