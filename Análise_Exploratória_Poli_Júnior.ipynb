## Questão 3
a) Crie um gráfico de latitude (lat) e longitude (long) dividido por distritos (neighbourhood group), o qual retornará o mapa de Nova York.

```python
from matplotlib.patches import Patch
df_2 = df.dropna(subset = ['lat','long'])
def transformar(bairro):
    if bairro == 'Bronx':
        a = 'red'
    elif bairro == 'Brooklyn':
        a = 'blue'
    elif bairro == 'Manhattan':
        a = 'green'
    elif bairro == 'Queens':
        a = 'purple'
    elif bairro == 'Staten Island':
        a = 'yellow'
    else:
        a = 'black'
    return a
lista = [ Patch(color='green', label='Manhattan'),  Patch(color='blue', label='Brooklyn'),  Patch(color='purple', label='Queens'),  Patch(color='red', label='Bronx'),  Patch(color='yellow', label='Staten Island')]
consertar = {'brookln':'Brooklyn','manhatan':'Manhattan'}
zeta = df_2['neighbourhood group'].replace(consertar)
df_2['neighbourhood group'] = zeta
df_2['cor'] = zeta.apply(transformar)
plt.scatter(df_2['long'],df_2['lat'],label = [df_2['neighbourhood group'].value_counts()], color = df_2['cor'], s= 0.1)
plt.title('Gráfico de Latitude e Longitude de Nova York Dividido por Distritos')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(handles=lista, loc='upper left', bbox_to_anchor=(1, 0.5))
plt.show()
```

b) Crie uma nova coluna que classifique os imóveis em 4 grupos de acordo com seus preços. A divisão deve ser baseada nos quartis, sendo que, por exemplo, até o valor limite do primeiro quartil de preço, os imóveis devem apresentar o valor "Q1" nessa nova coluna. Note que ao fim deveremos ter quatro classes de preços: Q1, Q2, Q3 e Q4; em proporçoes aproximadamente iguais na base.

```python
def trans(a):
    if a<= 340:
        b = 'Q1'
    elif (a <=624) and (a>340):
        b = 'Q2'
    elif (a <=913) and (a>624):
        b = 'Q3'
    else:
        b = 'Q4'
    return b
df['quartil'] = df.price.apply(trans)
print(df['quartil'])
```

## Etapa 2 - Estimativa de valor de mercado com Regressão

Para tal, o cliente pediu uma análise de regressão como forma de estimar o valor de mercado real das listagens a partir das vendas realizadas no passado na cidade, e comparar com os preços pedidos nas listagens.

De forma simples, um algoritmo de regressão é um modelo de Machine Learning capaz de prever um valor associado à um conjunto de variáveis. Isto é, a partir de uma base de treinamento com valores finais já fornecidos, o modelo aprende a indentificar características menos ou mais valiosas para o comportamento deste valor final, e utiliza isso para estimar um valor final para uma base que ainda não possui o valor definido.

Um exemplo simples pode ser o salário oferecido à jogadores de futebol:
- sabe-se que o jogador 1 teve 2 gols, 4 assistências, 72% de aproveitamento e recebeu 100 mil reais
- sabe-se que o jogador 2 teve 5 gols, 2 assistências, 64% de aproveitamento e recebeu 85 mil reais
- sabe-se que o jogador 3 teve 4 gols, 2 assistências, 67% de aproveitamento, quanto seria um salário justo para esse jogador?

Os modelos de regressão prometem fornecer uma resposta adequada para esta pergunta, levando em consideração tudo aquilo que foi analisado em sua base de teste, para prever um valor correspondente adequado.

A ideia da Poli Estates é utilizar a base de históricos de venda de imóveis da cidade para treinar o modelo, e utilizar toda a inteligência adquirida nesta etapa para identificar o valor de mercado das listagens atuais disponíveis no mercado. Se a listagem possuir um valor de mercado inferior ao pedido pelo proprietário, esta listagem pode ser considerada um mau investimento, porém, se o proprietário indicar um valor de venda inferior ao valor de mercado aferido pela análise de regressão, este pode ser considerado um bom investimento para a companhia.

Recomenda-se a utilização das diferentes opções de modelos de regressão implementados na biblioteca sklearn, e que sejam testados para descobrir a melhor performance para a base de dados estudada (recomenda-se o teste por validação cruzada). É importante ressaltar a necessidade de filtrar as características abordadas pelo modelo, de forma a não sobrecarregar o algoritmo com features (colunas utilizadas como varáveis independentes) dos imóveis que sejam pouco correlatas, ou capazes de gerar ruídos nas análises. Após o desenvolvimento e melhoria do modelo de regressão na base do histórico, aplicar o resultado sobre a base de vendas e identificar as melhores oportunidades de investimento para a companhia. Apresentar um relatório com o veredito sobre cada uma das propriedades ofertadas no mercado.

Para trazer esse veredito, basta aplicar a função predict do modelo que for treinado aos dados da base de vendas comentada. O erro deu consideravelmente baixo dentro dos dados que temos, agora vamos ver como ele se comporta quando criamos um data sample maior

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 42)
df_model = DecisionTreeRegressor(random_state=42)
df_model.fit(train_X,train_y)
preços_previstos_2 = df_model.predict(val_X)
print(mean_absolute_error(val_y,preços_previstos_2), preços_previstos_2.mean())
```

```python
fig,ax = plt.subplots()

ax.scatter(val_y,preços_previstos_2)

plt.show()
```

<span style="font-size:larger;">
Treinando uma base de dados maior, percebe-se que o erro aumenta.Agora utilizarei outro modelo de regressão

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 42)
def get_mae(max_folhas, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_folhas, random_state=42)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Numero máximo de nós %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

<span style="font-size:larger;">
50 é o numero de folhas ideal, isso quer dizer que o dataset não tem problemas de underfitting e overfitting e, portanto, tem uma boa precisão.Agora utilizaremos outro método para calcular o erro absoluto e a variação de y com o valor previsto

```python
df_sem = df.dropna()
df = df_sem
X = df[colunas_selec_2]
y = df.price
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 42)
modelo_arvore = RandomForestRegressor(random_state=1)
modelo_arvore.fit(train_X, train_y)
y_preds = modelo_arvore.predict(val_X)

fig,ax = plt.subplots()

ax.scatter(val_y,y_preds)

plt.show()
print(mean_absolute_error(val_y, y_preds),y_preds.mean())
```

<span style="font-size:larger;"> Esse modelo apresenta um menor mean absolute error e portanto é mais confiavel para uma previsão com um dataset maior 

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
num_clusters = 7
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
clusters = kmeans.fit_predict(X_scaled)
df['Cluster'] = clusters
plt.scatter(X['area'], y, c=clusters, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200)
plt.xlabel('price')
plt.ylabel('area')
plt.title('Resultados da Clusterização')
plt.show()
```
